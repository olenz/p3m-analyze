```python
%pylab inline
from sklearn.svm import SVR
import Q.q
```

```python
meshes, caos, alphas, Qik = Q.q.get()
def ixm(mesh): return searchsorted(meshes, mesh)
def ixc(cao): return searchsorted(caos, cao)
def ixa(alpha): return searchsorted(alphas, alpha)
Qik.min(), Qik.max()
```

```python
boundit = vectorize(lambda x: max(x, 1.e-14))
Qik = boundit(Qik)
Qik.min(), Qik.max()
```

```python
for cao in caos:
    loglog(alphas, Qik[ixm(512),ixc(cao),:])
```

```python
logQik = log10(Qik)
logQik.min(), logQik.max()
```

```python
def plotQ(mesh):
    meshix = searchsorted(meshes, mesh)
    for i in range(len(caos)):
        plot(alphas, logQik[meshix, i, :], label="cao={}".format(caos[i]))
    title("mesh={}".format(mesh))
    legend()
```

```python
plotQ(32)
```

```python
plotQ(64)
```

```python
plotQ(512)
```

```python
# create training vectors
mesh = 32
meshix = searchsorted(meshes, mesh)
Xs=[]
ys=[]
for caoi, cao in enumerate(caos):
    for alphai, alpha in enumerate(alphas):
            Xs.append([cao, alpha])
            ys.append(logQik[meshix, caoi, alphai])
X = array(Xs)
y = array(ys)
X.shape, y.shape
```

```python
# Train the SVR
svr = SVR(kernel='rbf', C=1000, epsilon=1.0)
fit = svr.fit(X,y)
print len(svr.support_vectors_)
```

```python
def plotQfit(mesh):
    for caoi, cao in enumerate(caos):
        plot(alphas, logQik[meshix, caoi, :], label="cao={}".format(cao))
        X = empty((len(alphas),2))
        X[:,0] = cao
        X[:,1] = alphas
        y = fit.predict(X)
        plot(alphas, y, '--')
    title("mesh={}".format(mesh))
    legend()
```

```python
plotQfit(32)
```
