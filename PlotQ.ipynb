```python
# import the data
import pickle
f = open('Q.pkl', 'r')
meshes, caos, alphas, Qik, Qiki, Qad, Qadi = pickle.load(f)
f.close()

# give a lower limit to the data
def boundit_single(x):
    if x < 1.e-13: return NaN
    else: return x
boundit = vectorize(boundit_single)
Qik = boundit(Qik)
Qiki = boundit(Qiki)
Qad = boundit(Qad)
Qadi = boundit(Qadi)

# throw away meshes < 8
meshes = meshes[2:]
Qiki = Qiki[2:,:,:]
Qadi = Qadi[2:,:,:]
Qad = Qad[2:,:,:]
Qik = Qik[2:,:,:]

def ixm(mesh): return searchsorted(meshes, mesh)
def ixc(cao): return searchsorted(caos, cao)
def ixa(alpha): return searchsorted(alphas, alpha)
```

```python
# a function that plots the different curves for a given mesh
def plotQ(mesh, Q):
    meshix = imx(mesh)
    for i in range(len(caos)):
        cao = caos[i]
        alphahs = alphas/meshes[meshix]
        nks = 1./alphahs
        subplot(121)
        #xlim((0.1, 20.0))
        loglog(alphahs, Q[meshix, i, :], '.', label="cao={}".format(cao))
        subplot(122)
        loglog(nks, Q[meshix, i, :], '.', label="cao={}".format(cao))
        #deriv = log10(Q[meshix, i, 1:]/Q[meshix, i, :-1]) / log10(alphas[1:]/alphas[:-1])
        #semilogx(alphas[1:], deriv, '.')

    subplot(121)
    title("mesh={}".format(mesh))
    xlabel(r'$\alpha h$')
    ylabel('Q')
    legend(loc="best")
    subplot(122)
    xlabel(r'$\frac{1}{\alpha h}$')
   
def plotAllQ(Q):
    for mesh in meshes:
        figure()
        plotQ(mesh, Q)
```

```python
# Master curve for IKI
for caoix, cao in enumerate(caos):
    figure(figsize=(15,4))
    for meshix, mesh in enumerate(meshes):
        alphahs = alphas/mesh
        #ylim((1.e-3, 1e2))
        subplot(121)
        loglog(alphahs, Qiki[meshix, caoix, :], '-')
        
        subplot(122)
        loglog(alphahs, Qiki[meshix, caoix, :]/mesh, '-')

    subplot(121)
    title("IKI cao={}".format(cao))
    xlim((1.e-1, 1.e1))
    ylim((1.e-10, 1.e2))
    xlabel(r"$\alpha h$")
    ylabel(r"$Q_{iki}$")
    
    subplot(122)
    xlim((1.e-1, 1.e1))
    ylim((1.e-10, 1.e2))
    xlabel(r"$\alpha h$")
    ylabel(r"$Q_{iki}/M$")
```

```python
# Master curve for ADI
for caoix, cao in enumerate(caos):
    figure(figsize=(15,4))
    for meshix, mesh in enumerate(meshes):
        alphahs = alphas/mesh
        #ylim((1.e-3, 1e2))
        subplot(121)
        loglog(alphahs, Qadi[meshix, caoix, :], '-')
        
        subplot(122)
        loglog(alphahs, Qadi[meshix, caoix, :]/mesh**1.35, '-')

    subplot(121)
    title("ADI cao={}".format(cao))
    xlim((1.e-1, 1.e1))
    ylim((1.e-10, 1.e2))
    xlabel(r"$\alpha h$")
    ylabel(r"$Q_{adi}$")
    
    subplot(122)
    xlim((1.e-1, 1.e1))
    ylim((1.e-10, 1.e2))
    xlabel(r"$\alpha h$")
    ylabel(r"$Q_{adi}/M$")
```
