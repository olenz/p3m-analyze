# Estimating $P^3M$ k-space errors: Universal error curve revisited

This work is a combination of the efforts of V. Ballenegger, F. Weik, H. Zahiri and O. Lenz.

The $P^3M$ algorithm has two variants which are of use in simulations: P3M-ad-i (analytically differntiated, interlaced) and P3M-ik-i (ik-differentiated, interlaced). In general, P3M-adi is the fastest (as shown in F. Weik's thesis), but it does not exactly conserve momentum, while P3M-iki does conserve momentum (up to numerical precision).

For all of these variants of the algorithm, accurate error estimate formulas exist (Deserno, Holm 1998) that can be used to get an upper limit of the rms error in the forces. These formulas depend on the following parameters:

* the number of charges in the system $N$
* the sum of the different squared charges in the system $\mathcal{Q}^2 = \sum q_i^2$
* the side length of the system $L$ (here we only consider cubic systems)
* the P3M cutoff distance $r_\mathrm{cut}$
* the P3M charge assignment order $P$
* the P3M mesh size $M$
* the P3M Ewald splitting parameter $\alpha$

These error estimates can be used to tune the P3M parameters. To get the best possible combination of parameters, the error estimate needs to be computed several times. Unfortunately, the formulas are still expensive to compute for large mesh sizes. Therefore, finding a cheaper way to estimate the errors is desireable.

Fortunately, we can show that the computationally expensive core $Q$ of the error estimate is independent of the system parameters $N$, $\mathcal{Q}^2$ and $L$ as well as the cutoff distance $r_\mathrm{cut}$, but does only depend on the P3M parameters $P$, $M$ and the dimensionless quantity $\alpha L$ (DA F. Weik). Therefore, we can precompute the values of $Q$ for all possible values of these parameters, and we do not need to know about specific properties of the system to estimate the error of the algorithms.

To be able to precompute $Q$, it is important to identify feasible limits of the different parameters for realistic systems. Here, we consider values of the CAO $P \le 7$ and mesh sizes of $8 \le M \le 512$. Larger mesh sizes are very expensive to compute.

The Ewald splitting parameter $\alpha$ is an inverse length, and we will call $r_\alpha = \frac{1}{\alpha}$ the *Ewald splitting radius*, where the near and the far field components of the potential are equal. We know that $r_\alpha$ should usually be smaller than the cutoff radius $r_\mathrm{cut}$, as otherwise we would not take into account a significant portion of the near field, but also that it should be not very much smaller, as in that case we would compute interactions in the near field that would not contribute to the result. If we consider the side length of a mesh cell $h = \frac{L}{M}$, we can define the quantity $r^*_\alpha = \frac{1}{\alpha h}$, which is the *reduced Ewald splitting radius* in units of the mesh cell size. Feasible values of $\alpha$ should be such that $0.1 < r^*_\alpha < 10.0$, which means that $0.1 < \alpha L < 10.0$.

## Postprocessing the data

Now we postprocess the data somewhat. First, we remove all values that are less than 1.e-13 because they are due to numerical errors. Then we remove all meshes smaller than 8, because they behave strangely and the error estimate doesn't really apply to these small meshes anyway.

```python
# give a lower limit to the data
def boundit_single(x):
    if x < 1.e-13: return NaN
    else: return x
boundit = vectorize(boundit_single)
Qik = boundit(Qik)
Qiki = boundit(Qiki)
Qad = boundit(Qad)
Qadi = boundit(Qadi)

# throw away meshes < 8
meshes = meshes[2:]
Qiki = Qiki[2:,:,:]
Qadi = Qadi[2:,:,:]
Qad = Qad[2:,:,:]
Qik = Qik[2:,:,:]

def ixm(mesh): return searchsorted(meshes, mesh)
def ixc(cao): return searchsorted(caos, cao)
def ixa(alpha): return searchsorted(alphas, alpha)
```

## Plotting the error curves

Now we can plot the data to get some feeling for it. First lets define a function to plot the error curve $\mathcal{Q}$ vs. $\alpha L$ for different values of $P$ and a given mesh $M$.

```python
# a function that plots the different curves for a given mesh
def plotQ(mesh, Q):
    meshix = ixm(mesh)
    for i in range(len(caos)):
        cao = caos[i]
        loglog(alphas, Q[meshix, i, :], '-', label="cao={}".format(cao))
    title("mesh={}".format(mesh))
    xlabel(r'$\alpha L$')
    ylabel('Q')
```

### Error curves of P3M-iki

```python
figure(figsize=(15,10))
subplot(221); plotQ(16, Qiki); legend(loc="best")
subplot(222); plotQ(64, Qiki)
subplot(223); plotQ(128, Qiki)
subplot(224); plotQ(512, Qiki)
```

### Error curves of P3M-adi

```python
figure(figsize=(15,10))
subplot(221)
plotQ(16, Qadi)
legend(loc="best")
subplot(222)
plotQ(64, Qadi)
subplot(223)
plotQ(128, Qadi)
subplot(224)
plotQ(512, Qadi)
```

## Master Curves

We see that the curves for a given value of the CAO $P$ but different values of the mesh $M$ look qualitatively similar. Therefore, let's try to find a master curve.

In the case of P3M-iki, it turns out that we can obtain a master curve for each value of the CAO $P$ by considering
$$Q^*_\mathrm{iki}(\alpha \frac{L}{M}, P) = \frac{Q_\mathrm{iki}(\alpha L, M, P)}{M}.$$

The following plots show the results of these master curves. In the left row, we plot $Q_\mathrm{iki}$ vs $\alpha L$ for a given CAO $P$ and all different values of the mesh $M$. In the right row, we plot $Q^*_\mathrm{iki}$ vs $\alpha h$. The result is pretty optimal.

All in all, to compute the error of the P3M-ik-i method, we can use the following formula: 
$$\Delta F = \mathcal{Q}^2 \sqrt{M \frac{Q^*_\mathrm{iki}(\alpha \frac{L}{M}, P)}{NL^3}}$$

Given that we can precompute $Q^*_\mathrm{iki}$, this allows the fast computation of the error for a given set of parameters.

```python
# Master curve for IKI
for caoix, cao in enumerate(caos):
    figure(figsize=(15,4))
    for meshix, mesh in enumerate(meshes):
        alphahs = alphas/mesh
        subplot(121)
        loglog(alphas, Qiki[meshix, caoix, :], '-')
        
        subplot(122)
        loglog(alphahs, Qiki[meshix, caoix, :]/mesh, '-')

    subplot(121)
    title("IKI cao={}".format(cao))
    xlabel(r"$\alpha L$")
    ylabel(r"$Q_\mathrm{iki}$")
    
    subplot(122)
    xlim((1.e-1, 1.e1))
    ylim((1.e-10, 1.e2))
    xlabel(r"$\alpha h$")
    ylabel(r"$Q^*_\mathrm{iki}$")
```

Now, let's try the same for P3M-ad-i. It turns out that the same formula as used in the case of P3M-ik-i does not work as good. Instead, a slightly modified formula seems to yield good results: 
$$Q^*_\mathrm{adi}(\alpha \frac{L}{M}, P) = \frac{Q_\mathrm{adi}(\alpha L, M, P)}{M^\frac{4}{3}}$$
and 
$$\Delta F = \mathcal{Q}^2 \sqrt{M^\frac{4}{3} \frac{Q^*_\mathrm{iki}(\alpha \frac{L}{M}, P)}{NL^3}}.$$

As the figures below show, the results are not a convincing as in the case of P3M-ik-i, but still we get a good estimate. 

Note the exponent of the mesh size of $\frac{4}{3}$! The exponent was found by trial and error. A good explanantion would be welcome. I assume that the exponent is connected to the error cancellation that makes the interlacing so effective in the case of P3M-ad.

```python
# Master curve for ADI
for caoix, cao in enumerate(caos):
    figure(figsize=(15,4))
    for meshix, mesh in enumerate(meshes):
        alphahs = alphas/mesh
        #ylim((1.e-3, 1e2))
        subplot(121)
        loglog(alphas, Qadi[meshix, caoix, :], '-')
        
        subplot(122)
        loglog(alphahs, Qadi[meshix, caoix, :]/mesh**(4./3.), '-')

    subplot(121)
    title("ADI cao={}".format(cao))
    xlabel(r"$\alpha L$")
    ylabel(r"$Q_\mathrm{adi}$")
    
    subplot(122)
    xlim((1.e-1, 1.e1))
    ylim((1.e-10, 1.e2))
    xlabel(r"$\alpha h$")
    ylabel(r"$Q^*_\mathrm{adi}$")
```
