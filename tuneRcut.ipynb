```python
%pylab inline
from scipy.optimize import curve_fit
```

```python
import pickle, os.path
#fname = 'tuneRcut/cloud_wall_8100/data.pkl'
#fname = 'tuneRcut/cloud_wall_8100_adi/data.pkl'
fname = 'tuneRcut/cloud_wall_102900/data.pkl'
#fname = 'tuneRcut/cloud_wall_102900_4cores/data.pkl'
#fname = 'tuneRcut/silica_melt_12960/data.pkl'
#fname = 'tuneRcut/silica_melt_103680/data.pkl'
with open(fname, 'r') as f:
    data = pickle.load(f)
near = data['near']
far = data['far']
total = data['total']
```

```python
figure(figsize=(15,8))
plot(total[:,0], total[:,1], '.', label="total")
plot(far[:,0], far[:,1], '.', label="far")
plot(near[:,0], near[:,1], '.', label="near")
legend()
```

```python
tmin = min(total[:,1])
rmin = total[argmin(total[:,1]),0]
print("rmin={} tmin={}".format(rmin,tmin))
```

```python
# Fit near field to c*r**3
def fnear(r, c): return c*r**3
pnear = curve_fit(fnear, near[:,0], near[:,1])
rs = linspace(min(near[:,0]), max(near[:,0]))

# Fit far field to c*r**(-3)
def ffar(r, c): return c*r**(-3)
pfar = curve_fit(ffar, far[:,0], far[:,1])
rs = linspace(min(far[:,0]), max(far[:,0]))

figure(figsize=(15,8))
plot(total[:,0], total[:,1], '.', label="total")
plot(far[:,0], far[:,1], '.', label="far")
plot(near[:,0], near[:,1], '.', label="near")
print "pnear =", pnear[0]
plot(rs, fnear(rs, *pnear[0]), '-', label="near fit")
print "pfar =",pfar[0]
plot(rs, ffar(rs, *pfar[0]), '-', label="far fit")
fest = lambda rs: fnear(rs, *pnear[0]) + ffar(rs, *pfar[0])
plot(rs, fest(rs), '-', label="fit sum")
legend()
```

```python
# Test prediction by fit
rnew = []
#tshift = 0.5*(pnear[0][1]+pfar[0][1])
for i in xrange(near.shape[0]):
    # Predict minimum for datapoint
    rcut, tnear = near[i]
    _, tfar = far[i]
    cnear = (tnear)/rcut**3
    cfar = (tfar)*rcut**3
    rmin = (cfar/cnear)**(1./6.)
    rnew.append([rcut,rmin])
rnew = array(rnew)

figure(figsize=(15,8))
ax = subplot(121)
ylabel('New $r_{cut}$')
xlabel('Initial $r_{cut}$')
plot(rnew[:,0], rnew[:,1], '.')
ax.set_aspect('equal', 'datalim')

# the indices of the closest real timings
ixs = searchsorted(total[:,0], rnew[:,1])
ts = map(lambda ix: min(total[ix:ix+4,1]), ixs)

subplot(122)
title("Relative time compared to optimal time when initial point is r")
plot(rnew[:,0], (ts-tmin)/tmin, '.')
xlabel("r")
ylabel("slowdown after 1 step")
```
